```python
import pandas as pd
import nest_asyncio
from telethon import TelegramClient
from telethon.sessions import StringSession
from telethon.tl.functions.messages import GetHistoryRequest
from telethon.tl.types import PeerChannel

# Apply asyncio patch to allow nested event loops in Jupyter
nest_asyncio.apply()

# Define API credentials
api_id = '28036061'
api_hash = '2400c765921b76f7cd50658053f0639b'
phone = '+251974619952'

# Initialize the Telegram client with an in-memory session
client = TelegramClient(StringSession(), api_id, api_hash)

keywords = [
    '#bank', '#CBE', '#BOA', '#GBE', '#CBO', '#ABAY', '#ABSC', '#ABYS', '#DASH', '#ENAT', '#BUNA',
    '#Commercial Bank of Ethiopia', '#bankofabyssinia', '#GlobalBankEthiopia', '#BoAEth', '#Coopbank', '#berhanbank'
]

async def main():
    await client.start(phone=phone)
    
    channel_username = 'tikvahethiopia'
    try:
        channel = await client.get_entity(channel_username)
    except Exception as e:
        print(f"Error: {e}")
        await client.disconnect()
        return

    def contains_keyword(message, keywords):
        if message and 'message' in message:
            message_text = message['message']
            for keyword in keywords:
                if keyword.lower() in message_text.lower():
                    return True
        return False

    async def get_all_messages(client, channel, keywords):
        all_messages = []
        offset_id = 0
        limit = 100
        total_messages = 0
        total_count_limit = 1000  # or some other limit you want to set

        while True:
            history = await client(GetHistoryRequest(
                peer=channel,
                offset_id=offset_id,
                offset_date=None,
                add_offset=0,
                limit=limit,
                max_id=0,
                min_id=0,
                hash=0
            ))
            if not history.messages:
                break
            messages = history.messages
            for message in messages:
                if contains_keyword(message.to_dict(), keywords):
                    all_messages.append(message.to_dict())
            offset_id = messages[-1].id
            total_messages = len(all_messages)
            if total_count_limit and total_messages >= total_count_limit:
                break

        return all_messages

    messages = await get_all_messages(client, channel, keywords)

    # Convert to DataFrame
    df = pd.DataFrame(messages)
    df.to_csv('telegram_channel_data.csv', index=False)

    print("Data extraction completed successfully")
    await client.disconnect()

# Run the main function
import asyncio
asyncio.run(main())
```

```python
def extract_attributes(row):
    # Define the keywords
    keywords = [
        '#bank', '#CBE', '#BOA', '#GBE', '#CBO', '#ABAY', '#ABSC', '#ABYS', '#DASH', '#ENAT', '#BUNA',
        '#Commercial Bank of Ethiopia', '#bankofabyssinia', '#GlobalBankEthiopia', '#BoAEth', '#Coopbank', '#berhanbank'
    ]
    
    # Extract message text
    message = row.get('message', '')
    
    # Find all keywords in the message
    keyword_found = [kw for kw in keywords if kw in message]
    keyword_found_str = ', '.join(keyword_found) if keyword_found else None

    # Extract post ID
    post_id = row.get('id', '')
    
    # Construct post link
    post_link = f"https://t.me/tikvahethiopia/{post_id}"
    
    # Extract views
    views = row.get('views', '')
    
    # Extract date and time
    date = pd.to_datetime(row['date']).date() if 'date' in row else ''
    post_hour = pd.to_datetime(row['date']).time() if 'date' in row else ''

    return pd.Series([keyword_found_str, post_link, views, date, post_hour])

# Apply the function to the DataFrame
df[['keyword', 'post_link', 'views', 'date', 'post_hour']] = df.apply(extract_attributes, axis=1)

# Select only the relevant columns
df_filtered = df[['keyword', 'post_link', 'views', 'date', 'post_hour']]

# Display the first few rows of the filtered DataFrame
print("First few rows of the updated DataFrame:")
print(df_filtered.head())
```

```python
import pandas as pd

# Load the CSV data
df = pd.read_csv('./telegram_channel_data.csv')

# Display the first few rows to understand the structure
print("First few rows of the original DataFrame:")
print(df.head())

def get_time_of_day(hour):
    """
    Determine the time of day based on the hour.
    """
    if 5 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 17:
        return 'afternoon'
    elif 17 <= hour < 21:
        return 'evening'
    else:
        return 'night'

def extract_attributes(row):
    """
    Extract specific attributes from each row of the DataFrame.
    """
    # Define the keywords to search for in the message
    keywords = [
        '#bank', '#CBE', '#BOA', '#GBE', '#CBO', '#ABAY', '#ABSC', '#ABYS', '#DASH', '#ENAT', '#BUNA',
        '#Commercial Bank of Ethiopia', '#bankofabyssinia', '#GlobalBankEthiopia', '#BoAEth', '#Coopbank', '#berhanbank'
    ]
    
    # Extract message text
    message = row.get('message', '')
    
    # Find all keywords in the message
    keyword_found = [kw for kw in keywords if kw in message]
    keyword_found_str = ', '.join(keyword_found) if keyword_found else None

    # Extract post ID
    post_id = row.get('id', '')
    
    # Construct post link
    post_link = f"https://t.me/tikvahethiopia/{post_id}"
    
    # Extract views
    views = row.get('views', '')
    
    # Extract date and time
    date = pd.to_datetime(row['date']).date() if 'date' in row else ''
    post_hour = pd.to_datetime(row['date']).time() if 'date' in row else ''
    
    # Calculate time_of_day
    hour = int(post_hour.hour) if post_hour else 0
    time_of_day = get_time_of_day(hour)

    return pd.Series([keyword_found_str, post_link, views, date, post_hour, time_of_day])

# Apply the function to the DataFrame
df_extracted = df.apply(extract_attributes, axis=1)

# Rename the columns for clarity
df_extracted.columns = ['keyword', 'post_link', 'views', 'date', 'post_hour', 'time_of_day']

# Display the first few rows of the extracted DataFrame
print("First few rows of the extracted DataFrame:")
print(df_extracted.head())

# Filter rows that contain "https://t.me/tikvahethiopia/" in the post_link column
filtered_df = df_extracted[df_extracted['post_link'].str.contains("https://t.me/tikvahethiopia/")]

# Display the first few rows of the filtered DataFrame
print("First few rows of the filtered DataFrame:")
print(filtered_df.head())

# Save the filtered DataFrame to a new CSV file
filtered_df.to_csv('filtered_telegram_data.csv', index=False)

# Display the total number of rows after filtering
print(f"Total rows after filtering: {len(filtered_df)}")
```
