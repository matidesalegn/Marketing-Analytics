```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import psycopg2

def get_time_of_day(hour):
    if 5 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 17:
        return 'afternoon'
    elif 17 <= hour < 21:
        return 'evening'
    else:
        return 'night'

class PostgresConnection:
    def __init__(self, dbname, user, password, host='localhost', port='5432'):
        self.dbname = dbname
        self.user = user
        self.password = password
        self.host = host
        self.port = port
        self.conn = None
        self.cursor = None

    def connect(self):
        try:
            self.conn = psycopg2.connect(
                dbname=self.dbname,
                user=self.user,
                password=self.password,
                host=self.host,
                port=self.port
            )
            self.cursor = self.conn.cursor()
            print("Connected to PostgreSQL database!")
        except Exception as e:
            print(f"Error: {e}")

    def create_table(self):
        create_table_query = """
        CREATE TABLE IF NOT EXISTS telegram_posts (
            id SERIAL PRIMARY KEY,
            date DATE NOT NULL,
            post_link TEXT NOT NULL,
            view TEXT,
            post_hour TIME,
            bank TEXT,
            time_of_day TEXT
        );
        """
        try:
            self.cursor.execute(create_table_query)
            self.conn.commit()
            print("Table 'telegram_posts' created successfully!")
        except Exception as e:
            print(f"Error creating table: {e}")
            self.conn.rollback()

    def insert_data(self, df, table_name):
        if self.conn is not None:
            try:
                for i, row in df.iterrows():
                    self.cursor.execute(f"""
                        INSERT INTO {table_name} (date, post_link, view, post_hour, bank, time_of_day)
                        VALUES (%s, %s, %s, %s, %s, %s)
                    """, tuple(row))
                self.conn.commit()
                print("Data inserted successfully!")
            except Exception as e:
                print(f"Error: {e}")
                self.conn.rollback()

    def close_connection(self):
        if self.conn is not None:
            self.cursor.close()
            self.conn.close()
            print("Connection closed.")

def extract_data(post_link, bank):
    # Initialize a browser driver (e.g., Chrome)
    driver = webdriver.Chrome()

    # Load the Telegram post URL
    driver.get(post_link)

    # Wait for the iframe to load (adjust the wait time if needed)
    driver.implicitly_wait(10)  # Wait up to 10 seconds

    # Get the page source after JavaScript execution
    html = driver.page_source

    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Find the iframe element
    iframe = soup.find('iframe')
    if not iframe:
        print(f"Error: No iframe found for post {post_link}")
        driver.quit()
        return None

    # Get the iframe source URL
    iframe_src = iframe['src']

    # Make a new request to the iframe source
    iframe_response = requests.get(iframe_src)
    iframe_soup = BeautifulSoup(iframe_response.content, 'html.parser')

    # Now you can search for the desired elements within iframe_soup
    info_div = iframe_soup.find('div', class_='tgme_widget_message_info js-message_info')

    # Extract views
    views_span = info_div.find('span', class_='tgme_widget_message_views')
    views = views_span.text.strip() if views_span else None

    # Extract date and time
    date_element = info_div.find('a', class_='tgme_widget_message_date')
    datetime_element = date_element.find('time', class_='datetime') if date_element else None
    date_time = datetime_element['datetime'] if datetime_element else None

    # Close the browser driver
    driver.quit()

    if date_time:
        datetime_object = datetime.fromisoformat(date_time)
        date = datetime_object.strftime('%Y-%m-%d')  # Format date as YYYY-MM-DD
        post_hour = datetime_object.strftime('%H:%M')  # Format time as HH:MM
        time_of_day = get_time_of_day(int(post_hour.split(':')[0]))

        # Create the dictionary
        data = {
            'Views': views,
            'Date': date,
            'Post Hour': post_hour,
            'Bank': bank,
            'Time of Day': time_of_day,
            'Post Link': post_link  # Include the post link
        }

        print(data)
        return data
    else:
        print("Date and time not found in the iframe.")
        return None

# List of posts and associated banks
posts = [
    {'post_link': 'https://t.me/tikvahethiopia/86164', 'bank': 'CBE'},
    {'post_link': 'https://t.me/tikvahethiopia/79089', 'bank': 'CBE'},
    {'post_link': 'https://t.me/tikvahethiopia/87041', 'bank': 'GBE'},
    {'post_link': 'https://t.me/tikvahethiopia/86185', 'bank': 'GBE'},
    {'post_link': 'https://t.me/tikvahethiopia/87693', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/87496', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/87443', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/87344', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/87236', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/87018', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/86867', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/86831', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/86555', 'bank': 'BOA'},
    {'post_link': 'https://t.me/tikvahethiopia/87532', 'bank': 'BOA'}
]

# Extract data and insert into PostgreSQL
db = PostgresConnection(dbname='bank', user='postgres', password='Mati@1993')
db.connect()
db.create_table()  # Ensure the table exists before inserting data

for post in posts:
    extracted_data = extract_data(post['post_link'], post['bank'])
    if extracted_data:
        # Create DataFrame
        data = {
            'date': [extracted_data['Date']],
            'post_link': [extracted_data['Post Link']],
            'view': [extracted_data['Views']],
            'post_hour': [extracted_data['Post Hour']],
            'bank': [extracted_data['Bank']],
            'time_of_day': [extracted_data['Time of Day']]
        }
        df = pd.DataFrame(data)

        # Insert data into PostgreSQL
        db.insert_data(df, 'telegram_posts')

db.close_connection()
```
